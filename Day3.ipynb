```python
def count_test(n):
  count=1
  while count <=n:
    yield count
    count=count+1
```

```python
c=count_test(5)
```

```python
print(c)
```

```python
for i in c:
  print(i)
```

```python
n=12345
n=abs(n)
count =0
if n==0:
  count=1
else:
  while n>0:
    count+=1
    n//=10
print(count)
```

POLYMORPHISM

```python
class UPI:
  def pay(self):
      print("payment through UPI")
class Card:
    def pay(self):
      print("payment through Card")
for payment in [UPI(),Card()]:
     payment.pay()
```
