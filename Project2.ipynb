```python
def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 40:
        return "D"
    else:
        return "E"

print("=====Student Marksheet====")
name= input("Enter student name:")
subject1=int(input("Enter math marks:"))
subject2=int(input("Enter English marks:"))
subject3=int(input("Enter Physics marks:"))
total_marks=subject1+subject2+subject3
result=(total_marks/300)*100
print(f"your cgpa is{result/9.5}")

print("math Marks and Grade:",subject1, grade(subject1))
print("English Marks and Grade:",subject2, grade(subject2))
print("Physics Marks and Grade:",subject3, grade(subject3))




```
